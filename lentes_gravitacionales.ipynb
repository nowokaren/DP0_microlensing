```python
import matplotlib.pyplot as plt
import numpy as np

# Parámetros de diagrama
D_perp = 1.0  # Distancia entre los puntos de observación en unidades arbitrarias
t_E = 20.0    # Tiempo de Einstein en días
Delta_t0 = 5.0 # Diferencia en tiempo de máximo acercamiento
Delta_u0 = 0.1 # Diferencia en el parámetro de impacto

# Configuración del gráfico
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(-1, 2)
ax.set_ylim(-0.5, 1.5)

# Dibujar la línea de visión y los puntos de observación
ax.plot([0, 0], [-0.5, 1.5], 'k--', label='Línea de visión')
ax.plot([-0.5, 0.5], [1, 1], 'ro', markersize=10, label='Puntos de observación')

# Dibujar el vector de paralaje
ax.arrow(0, 0.5, Delta_t0/(t_E * D_perp), Delta_u0/D_perp, 
         head_width=0.05, head_length=0.05, fc='blue', ec='blue')
ax.text(Delta_t0/(2 * t_E * D_perp), Delta_u0/(2 * D_perp), 
        r'$\vec{\pi}_E$', color='blue', fontsize=12)

# Etiquetas de los elementos
ax.text(-0.05, -0.2, r'Lente y fuente alineadas', ha='center', fontsize=10)
ax.text(-0.5, 1.1, r'Observador 1', color='red', fontsize=10)
ax.text(0.5, 1.1, r'Observador 2', color='red', fontsize=10)
ax.text(-0.8, 0.8, r'$D_\perp$', fontsize=12, rotation=0)

# Título y leyendas
plt.title("Diagrama de paralaje en un evento de microlente")
plt.legend(loc=(1.01,0))
plt.xlabel('Dirección de la lente')
plt.ylabel('Dirección perpendicular')
plt.xlim(-1,1)
plt.ylim(-0.25,1.25)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(6,6))
# Definimos los puntos clave en el diagrama
source = np.array([3, 10])  # Posición de la fuente
lens = np.array([3, 5])     # Posición de la lente
obs1 = np.array([1, 1])     # Observador 1
obs2 = np.array([9, 1])     # Observador 2

# Distancia perpendicular entre los observadores
D_perp_line_x = [obs1[0], obs2[0]]
D_perp_line_y = [obs1[1], obs2[1]]

# Línea de alineación entre lente y fuente
lens_source_line_x = [lens[0], source[0]]
lens_source_line_y = [lens[1], source[1]]

# Añadimos etiquetas para las distancias
plt.text(5,1.5, r'$D_\perp$', color='k', fontsize=14, ha='center')

# Líneas entre observadores y la lente
obs1_lens_line_x = [obs1[0], lens[0]]
obs1_lens_line_y = [obs1[1], lens[1]]
obs2_lens_line_x = [obs2[0], lens[0]]
obs2_lens_line_y = [obs2[1], lens[1]]

# Líneas entre observadores y la fuente
obs1_source_line_x = [obs1[0], source[0]]
obs1_source_line_y = [obs1[1], source[1]]
obs2_source_line_x = [obs2[0], source[0]]
obs2_source_line_y = [obs2[1], source[1]]

# Cálculo del punto de impacto (proyección perpendicular) para cada observador
def perpendicular_point(lens, obs, source):
    # Vector de observador a fuente
    obs_source_vec = source - obs
    # Vector de lente a observador
    lens_obs_vec = lens - obs
    # Proyección de la lente sobre la línea observador-fuente
    proj_length = np.dot(lens_obs_vec, obs_source_vec) / np.linalg.norm(obs_source_vec)**2
    proj_point = obs + proj_length * obs_source_vec
    return proj_point

# Proyecciones perpendiculares (parámetros de impacto) desde la lente hacia la línea observador-fuente
impact_point_obs1 = perpendicular_point(lens, obs1, source)
impact_point_obs2 = perpendicular_point(lens, obs2, source)

# Líneas del parámetro de impacto para cada observador
impact_line_obs1_x = [lens[0], impact_point_obs1[0]]
impact_line_obs1_y = [lens[1], impact_point_obs1[1]]
impact_line_obs2_x = [lens[0], impact_point_obs2[0]]
impact_line_obs2_y = [lens[1], impact_point_obs2[1]]

# Dibujamos el parámetro de impacto como líneas punteadas
plt.plot(impact_line_obs1_x, impact_line_obs1_y, 'g--', label='Parámetro de impacto desde $O_1$')
plt.plot(impact_line_obs2_x, impact_line_obs2_y, 'b--', label='Parámetro de impacto desde $O_2$')

# Distancia perpendicular y línea de alineación
plt.plot(D_perp_line_x, D_perp_line_y, 'k--', label=r'$D_\perp$ (Distancia entre observadores)')
plt.plot(lens_source_line_x, lens_source_line_y, 'k--', label='Línea de alineación Lente-Fuente')

# Líneas adicionales hacia la lente y la fuente
plt.plot(obs1_lens_line_x, obs1_lens_line_y, 'g:', label='Línea O1-L')
plt.plot(obs2_lens_line_x, obs2_lens_line_y, 'b:', label='Línea O2-L')
plt.plot(obs1_source_line_x, obs1_source_line_y, 'g:', label='Línea O1-S')
plt.plot(obs2_source_line_x, obs2_source_line_y, 'b:', label='Línea O2-S')

# Puntos en el diagrama
plt.scatter(*source, color='purple', s=100, label="Fuente (S)")
plt.scatter(*lens, color='orange', s=100, label="Lente (L)")
plt.scatter(*obs1, color='green', s=100, label="Observador 1 ($O_1$)")
plt.scatter(*obs2, color='blue', s=100, label="Observador 2 ($O_2$)")

# Etiquetas de los ángulos en posiciones aproximadas
plt.text(1.75, 3, r'$\theta_1$', color='green', fontsize=14, ha='center')
plt.text(7.5, 2.3, r'$\theta_2$', color='blue', fontsize=14, ha='center')

plt.text(2.5, 5.5, r'$u_0$', color='green', fontsize=14, ha='center')
plt.text(4,6, r'$u_0$', color='blue', fontsize=14, ha='center')

# Configuramos los ejes y el resto del diagrama
plt.xlim(0, 10)
plt.ylim(0, 11)
plt.xlabel("Posición en X")
plt.ylabel("Posición en Y")
plt.legend(loc=(1.01,0))
plt.grid(True)
plt.savefig("parallax.png")
plt.show()
```
