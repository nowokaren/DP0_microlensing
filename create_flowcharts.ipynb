```python
# Run.generate_location()

from graphviz import Digraph

def generate_flowchart():
    dot = Digraph()
    dot.attr(size='12,12')
    
    # Nodos
    dot.node('A', '¿dist es None?')
    dot.node('B', '¿calexps_method es "htm"?')
    dot.node('C', 'ref_dist = triangle_min_height(htm_vertex)')
    dot.node('D', 'ref_dist = radius')
    dot.node('E', 'dist = ref_dist / 20')
    dot.node('F', 'dist = dist')
    dot.node('G', '¿calexps_method es "htm"?')
    dot.node('H', 'ra, dec = tri_sample(htm_vertex)')
    dot.node('I', 'ra, dec = circ_sample(ra, dec, radius, margin=dist)')
    dot.node('J', '¿sources es None?')
    dot.node('K', 'sources = load_sources()')
    dot.node('L', '¿blend es None?')
    dot.node('M', 'Cargar fuentes cercanas y verificar blend')
    dot.node('N', 'Determinar si inyectar blend')
    dot.node('O', 'Recursión con generate_location')
    dot.node('P', 'Calcular nueva posición con blend')
    dot.node('Q', 'Retornar coordenadas blend')
    dot.node('R', 'Cargar fuentes cercanas sin blend')
    dot.node('S', 'Verificar si hay fuente cercana (<0.002°)')
    dot.node('T', 'Recursión con generate_location')
    dot.node('U', 'Verificar si hay evento cercano (<0.001°)')
    dot.node('V', 'Recursión con generate_location')
    dot.node('W', 'Retornar coordenadas (ra, dec, None, None)')
    
    # Conexiones
    dot.edge('A', 'B', label='Sí')
    dot.edge('A', 'F', label='No')
    dot.edge('B', 'C', label='Sí')
    dot.edge('B', 'D', label='No')
    dot.edge('C', 'E')
    dot.edge('D', 'E')
    dot.edge('E', 'G')
    dot.edge('F', 'G')
    dot.edge('G', 'H', label='Sí')
    dot.edge('G', 'I', label='No')
    dot.edge('H', 'J')
    dot.edge('I', 'J')
    dot.edge('J', 'K', label='Sí')
    dot.edge('J', 'L', label='No')
    dot.edge('K', 'L')
    dot.edge('L', 'M', label='No')
    dot.edge('L', 'R', label='Sí')
    dot.edge('M', 'N')
    dot.edge('N', 'O', label='No hay fuentes')
    dot.edge('N', 'P', label='Sí hay fuentes')
    dot.edge('O', 'M')
    dot.edge('P', 'Q')
    dot.edge('Q', 'W')
    dot.edge('R', 'S')
    dot.edge('S', 'T', label='Sí')
    dot.edge('S', 'U', label='No')
    dot.edge('T', 'W')
    dot.edge('U', 'V', label='Sí')
    dot.edge('U', 'W', label='No')
    dot.edge('V', 'W')
    
    return dot


diagram = generate_flowchart()
diagram.render('run_generate_location', format='png', view=True)
```
