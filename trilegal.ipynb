```python
import requests

url = "https://stev.oapd.inaf.it/cgi-bin/trilegal"

response = requests.get(url)

if response.status_code == 200:
    print("Página cargada exitosamente.")
    
    print(response.text[:500]) 
else:
    print(f"Error al cargar la página. Código de estado: {response.status_code}")
```

```python
from bs4 import BeautifulSoup

if response.status_code == 200:
    print("Page loaded.")
    soup = BeautifulSoup(response.text, 'html.parser')

    # Search form and input fields
    forms = soup.find_all('form')
    inputs = soup.find_all('input')

    print(f"Finded forms: {len(forms)}")
    print("{:20} {:10} {:10}".format("Field", "Type", "Default"))
    for input_field in inputs:
        name = input_field.get('name')
        type_field = input_field.get('type')
        value = input_field.get('value', '')
        print("{:20} {:10} {:10}".format(name, type_field, value))
else:
    print(f"Error loading page: {response.status_code}")
```
